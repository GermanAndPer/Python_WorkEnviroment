# ✅ Buenas Practicas

## ⬇️💻Bases Pesadas

En caso de trabajar en grandes proyectos, es mejor trabajar en ambientes separados.

No un ambiente por proyecto. Ir mas lejos y tener un ambiente para cada caso.



Por lo que se recomienda tener una carpeta donde se tengan todos los ambientes.
