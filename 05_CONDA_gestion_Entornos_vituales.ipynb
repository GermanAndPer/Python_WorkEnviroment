# üêç Conda: crear, actualizar y eliminar ambientes

Para listar los ambientes en la computadora, se utiliza el siguiente comando.

```python
!conda env list
```

El que se llama base, es el que Anaconda crea por Default

indicando con un * el ambiente donde estamos



El siguiente codigo es para crear un ambiente.

En el podemos especificar todas las librerias que necesitemos y sus versiones.

O podemos dejar solo Python y agregar las demas manualmente.

```python
!conda create --name py35 python pandas
#--name py35  es para crear un entorno virtual llamado py35. (Se puede cambiar "py35" pore cualauier otro)
#python       es para instalar la version mas reciente de python
#python3.5    es para instalar la version 3.5 de python
#pandas       es para instalar la libreria mas reciente de pandas. Pero se puede asignar cualqquie otra
```

Una vez creado al final del texto de instalacion, nos apacecer√° el siguiente codigo.

Siendo el primero para activar el ambiente y el segundo para desactivarlos.



```python
!conda activate py35
!conda deactivate
```

Permitiendonos as√≠ pasar de un amvbiente a otro. Lo podremos ver actualizando la leyenda de "base" al principio.



Una vez situado desde cualquiera de los de los ambientes, para saber cuales con las librerias instaladas en dicho ambiente se utiliza el siguiente comando.

```python
!conda list
```



En caso de querer ver solo una librera en especifica, se usa el mismo comando.

Pero se agrega el nombre de la libreria:

```python
!conda list python
```



Para actualizar una liberia usamos el siguiente codigo.

```python
!conda update python
#update para indicar que actualizaremos una libreria. Depues el nombre de la libreria.
!conda update pandas
```

En caso de que no haberla instalado, usamos el siguiente codigo para haccerlo. 

```python
!conda install pandas
#Tambien se pude especificar la version. O la mas actual.
```

En caso de no ser compatibles la version de python y la de pandas que se quiera instalar, aparecer√° un mensaje sugiriendo hacer una u otra modificacion.



Y en caso de necesitarlo, podemos hacer todas las instalaciones necesarias al mismo tiempo. 

Por ejemplo:

```python
!conda install python=3.9 pandas=1.2
```

## üóíÔ∏è Renombrado

Para renombrar actualmente no hay una funcion. Por lo que se har√° una copia. con el nombre correcto.

Usamos el siguiente codigo:

```python
!conda create --name py3_10_18 --copy --clone py35
#--copy      capia todo el contenido del ambiente
#--clone     crea un ambiente nuevo con el mismo contenido que el ambiente original
#--name      nombre del ambiente
#py35        nombre del ambiente a copiar
```

## üóëÔ∏èElimiar ambientes

```python
!conda env remove --name py35
#py35 es el nombre del ambiente
#No se puede eliminar el ambiente mientras estas en el. Debes cambiar a otro
```
