Mamba es una re interpretacion de Conda pero en C++

Por lo que es muchisimo mas rapida.

El codigo para activarla instalarla es:

```python
!conda install --channel conda-forge mamba
```

Mamaba se utiliza igual que CONDA. Solo usando el comando MAMBA

```python
!mamba --help
```
