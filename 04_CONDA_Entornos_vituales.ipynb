# 🐍CONDA🐍 

CONDA es una herramienta que nos ayuda a genera entornos virtuales

##  💡¿Por que es importante?

- Permite trabajar en diferentes proyectos sin conflictos de dependencias.
- Facilita la instalación de paquetes y dependencias.
- Permite crear entornos de desarrollo reproducibles.
- Permite trabajar en diferentes versiones de paquetes.
- Permite trabajar en diferentes sistemas operativos.
- Permite trabajar en diferentes lenguajes de programación.

Estas son los principales beneficios de CONDA

Permite generar diferentes proyectos y cada uno de estos proyectos pueden trabajar con diferentes ambientes.

Permitidiendo así trabajar en diferentes proyectos y que cada uno de estos trabaje con diferentes versiones de Python u otras caracteristicas sin que unas afecten a otras.

Permitiendo no solo trabajar con 2 ambientes, si no con muchas. Sin importar la necesidad.

 

## 💡¿Con que lenguajes de programacion funciona?

- C

## 🐍 Versiones de Conda

Importante mencionar que hay carias opciones.

- __Conda__ : siendo la version mas ligera
- __MiniConda__: Una version que tiene solo lo indispensable.
- __Anaconda__: La version completa.
