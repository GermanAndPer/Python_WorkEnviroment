# ðŸ” Diferentes versiones

En caso de que hayas instalado alguna libreria o la hayas instalado y genere conflicto con las otras librerias, siempre podemos volver al punto anterior.

Seria con apoyo de las __Revisiones__

Las revisiones son los estados en los que se encuentra el ambiente. Incluyendo las librerias y sus versiones.

Por lo que cada que se instala una nueva o se actualiza se genera una nueva version.

Para ver las versiones existentes se utiliza el comando:

```python
!conda list --revision
```

El numero junto al REV es el nmumero de la version que se necesita



Para actualizarlo necesitamos el siguiente codigo.

```python
!conda install --revision 0
# 0     Siendo la version de la instalacion
```
